# How To: Convert to Datetime Notation

First, let's create a new dataframe with the begin and end date data.

Let's see what kind of object the begin and end date columns are.

```python
costly_disasters.info()
```

It's an int64 type so we can manipulate it using modular arithmetic to get year, month, and day data, then turn that into a dataframe.

```python
begin_date = pd.DataFrame({'year':costly_disasters['Begin Date']//10000,'month':costly_disasters['Begin Date']//100%100,'day':costly_disasters['Begin Date']%100})
end_date = pd.DataFrame({'year':costly_disasters['End Date']//10000,'month':costly_disasters['End Date']//100%100,'day':costly_disasters['End Date']%100})
```

Let's convert to DateTime notation:

```python
begin_date_list = pd.to_datetime(begin_date)
end_date_list = pd.to_datetime(end_date)
```

Let's replace the columns in the original dataframe with these new ones.

```python
costly_disasters['Begin Date']=begin_date_list
costly_disasters['End Date']=end_date_list
```

```python
costly_disasters
```

We could also have converted the begin and end dates to timstamps in-place.

# How To: Convert to Datetime Notation (Again)

Let's import the dataset again since I overwrote the columns of interest in the last example.

```python
costly_disasters2 = pd.read_csv('/Users/nikita/Desktop/events-US-1980-2023.csv')
```

Let's go row by row.

```python
for i in range(len(costly_disasters2)):
    date = costly_disasters2['Begin Date'].iloc[i]
    end_date2 = costly_disasters2['End Date'].iloc[i]
    costly_disasters2['Begin Date'][i]=pd.Timestamp(year=date//10000,month=date//100%100,day=date%100)
    costly_disasters2['End Date'][i]=pd.Timestamp(year=end_date2//10000,month=end_date2//100%100,day=end_date2%100)
```

```python
costly_disasters2
```
